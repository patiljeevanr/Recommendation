```python
import pandas as pd
import numpy as np
```

```python
header = ['user_id', 'item_id', 'rating', 'timestamp']
dataset = pd.read_csv('data/ratings_Digital_Music.csv', names=header)
```

```python
dataset.head()
```

```python
df = dataset.head(10000)
```

```python
n_users = df.user_id.unique().shape[0]
n_items = df.item_id.unique().shape[0]
print ('Number of users = ' , str(n_users) , ' | Number of movies = ' , str(n_items))
```

```python
from sklearn import cross_validation as cv
train_data, test_data = cv.train_test_split(df, test_size=0.25)
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
train_data_matrix = train_data.pivot(index='user_id', columns='item_id', values='rating').fillna(0)
train_data_matrix.head()
```

```python
train_data_matrix.shape
```

```python
test_data_matrix = test_data.pivot(index='user_id', columns='item_id', values='rating').fillna(0)
test_data_matrix.head()
```

```python
test_data_matrix.shape
```

```python
from sklearn.metrics.pairwise import pairwise_distances
user_similarity = pairwise_distances(train_data_matrix, n_jobs=-1, metric='cosine')
item_similarity = pairwise_distances(train_data_matrix.T, n_jobs=-1, metric='cosine')
```

```python
user_similarity
```

```python
item_similarity
```

```python
def predict(ratings, similarity, type='user'):
    if type == 'user':
        mean_user_rating = ratings.mean(axis=1)
        #You use np.newaxis so that mean_user_rating has same format as ratings
        #ratings_diff = (ratings - mean_user_rating[:, np.newaxis])
        ratings_diff = ratings.sub(ratings.mean(axis=1), axis=0)
        pred = mean_user_rating[:, np.newaxis] + similarity.dot(ratings_diff) / np.array([np.abs(similarity).sum(axis=1)]).T
    elif type == 'item':
        pred = ratings.dot(similarity).div(pd.DataFrame(np.array([np.abs(similarity).sum(axis=1)])).iloc[0]).values
    return pred
```

```python
item_prediction = predict(train_data_matrix, item_similarity, type='item')
user_prediction = predict(train_data_matrix, user_similarity, type='user')
```

```python
item_prediction.shape
```

```python
test_data_matrix.shape
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
def rmse(prediction, ground_truth):
    prediction = prediction[ground_truth.nonzero()].flatten()
    ground_truth = ground_truth[ground_truth.nonzero()].flatten()
    return sqrt(mean_squared_error(ground_truth, prediction))

def mse(prediction, ground_truth):
    prediction = prediction[ground_truth.nonzero()].flatten()
    ground_truth = ground_truth[ground_truth.nonzero()].flatten()
    return mean_squared_error(ground_truth, prediction)
```

```python
print ('User-based CF MSE: ' , str(mse(user_prediction, test_data_matrix.values)))
print ('Item-based CF MSE: ' , str(mse(item_prediction, test_data_matrix.values)))
```

```python
print ('User-based CF RMSE: ' , str(rmse(user_prediction, test_data_matrix.values)))
print ('Item-based CF RMSE: ' , str(rmse(item_prediction, test_data_matrix.values)))
```

```python
sparsity=round(1.0-len(df)/float(n_users*n_items),3)
print ('The sparsity level of Music dataset 10000 is ' ,  str(sparsity*100) , '%')
```

```python
import scipy.sparse as sp
from scipy.sparse.linalg import svds

#get SVD components from train matrix. Choose k.
u, s, vt = svds(train_data_matrix, k = 20)
s_diag_matrix=np.diag(s)
X_pred = np.dot(np.dot(u, s_diag_matrix), vt)
print ('User-based CF MSE: ' , str(mse(X_pred, test_data_matrix.values)))
print ('User-based CF RMSE: ' , str(rmse(X_pred, test_data_matrix.values)))
```
